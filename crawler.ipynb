```python
import requests
url = 'https://sys.ndhu.edu.tw/SA/XSL_ApplyRWD/ActApply.aspx?lang=en'
html = requests.get(url)
print(html.encoding)
```

```python
from bs4 import BeautifulSoup

sp = BeautifulSoup(html.text, 'html.parser')
print(sp.title)
```

```python
table = sp.find('table', {'id': 'BodyContent_gvActs'})
print(table)
```

```python
headers = [th.text.strip() for th in table.find_all('th')]
print(headers)
```

```python
activity_names = []
for x in range(10):
    act = table.find('span', id=f"BodyContent_gvActs_lblGv_act_name_{x}").text
    activity_names.append(act)
print(activity_names)
```

```python
act = table.find('span', id=f"BodyContent_gvActs_lblGv_act_name_0")
print(type(act))
print(act.text)
```

```python
hours_certified = []
for x in range(10):
    num = int(table.find('span', id=f"BodyContent_gvActs_lblGv_xsl_check_{x}").text[-1])
    hours_certified.append(num)
print(hours_certified)
```

```python
num_people = []
for x in range(10):
    num = int(table.find('span', id=f"BodyContent_gvActs_lblGv_reg_num_{x}").text.split('/')[0].replace(' ', ''))
    num_people.append(num)
print(num_people)
```

```python
date_time_reg = []
for x in range(10):
    date_time = table.find('span', id=f"BodyContent_gvActs_lblGv_reg_dt_{x}").text
    date_time_reg.append(date_time)
print(date_time_reg)
```

```python
date_time_act = []
for x in range(10):
    date_time = table.find('span', id=f"BodyContent_gvActs_lblGv_act_dt_{x}").text
    date_time_act.append(date_time)
print(date_time_act)
```

```python
columns = ['Name of the activity', 'Number of hours certified', 'Number of people who have registered', 'Date and time of registration', 'Date and time of the activity']
rows = [activity_names, hours_certified, num_people, date_time_reg, date_time_act]

# print(rows[0])
import numpy as np
transposed = np.array(rows).T
print(transposed)
```

```python
import pandas as pd
df = pd.DataFrame(transposed, columns=columns)
df.index = np.arange(1, len(df) + 1)
from IPython.display import display
display(df)
```
